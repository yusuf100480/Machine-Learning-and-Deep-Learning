## About the dataset

This is perhaps the best known database to be found in the pattern recognition literature. The data set contains 3 classes of 50 instances each, where each class refers to a type of iris plant. Attribute Information:
   1. sepal length in cm
   2. sepal width in cm
   3. petal length in cm
   4. petal width in cm

### Find out the no of male & female employees in the company

*Hint: Use can use a condition to find out the no of 1s and 0s in gender and sum it*

```python
## The gender field contains 1 and 2. There is no 0 available in gender field. Assuming 1 is male and 2 is female.

# Find the unique values for Gender column
dfhr.Gender.unique()
```

```python
# Find the number of Male employees
male = len(dfhr[dfhr["Gender"] == 1])

# Find the number of female employees
female = len(dfhr[dfhr["Gender"] == 2])

print("Number of Male:",male)
print("Number of Female:",female)
print("Total no.of employees:", male+female)
```

### Plot the attrition column (Since it is a categorical variable, chose the plot accordingly)

```python
# Import matplotlib package to plot attrition column

import matplotlib.pyplot as plt

# Get the count for each categorical variable 
x = dfhr['Attrition'].value_counts()

# Plot the bar chart
x.plot(kind='bar',figsize=(10,5),color='blue')
plt.xlabel("Attrition")
plt.ylabel("Count")
plt.title('Number of Attrition')
plt.show()
```

### Plot Total working years using matplotlib plot function

```python
# Import matplotlib for plotting TotalWorkingYears
import matplotlib.pyplot as plt

# Get the groupby value count for TotalWorkingYears
s = dfhr.groupby(by="TotalWorkingYears", as_index=False).size()

# Set the chart size
plt.figure(figsize=(15,10))

# Plot the bar chart
s.plot.bar()
plt.xlabel("Total Working Years")
plt.ylabel("Count")
plt.show()

```
